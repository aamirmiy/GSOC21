### B.How are the lines in feature C related?And what causes the plunge near 900 Angstroms?

The lines in feature C belong to the Hydrogen spectrum belonging to the Lyman Series(911Angstrom to 1215Angstrom)


The plunge near 900 Angstrom is caused because the radiations approach the limit of Lyman series for Hydrogen. As n approaches infinity the spectral lines end up appearing to converge at this wavelength(911 Angstrom Lyman Limit) which causes radiations in these frequencies to be absorbed by the gas.Since it is being absorbed, a plunge can be seen in the graph.
